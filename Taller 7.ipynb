```python
%pylab inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

Crea el vector x con las posiciones dentro del intervalo entre 0 y 2.
El vector IC de condicion inicial para u

```python
x = linspace(0,2,200)
IC = 3.0*sin(pi*x)
```

dx es la separacion en x </br>
dt es la separacion temporal </br>
beta es el numero de Courant-Friederichs-Levy

```python
dx = 2.0/199
dt = 0.001
beta = dt/dx
```

Se define el numero de iteraciones basados en dt </br>
Se crea el vector de tiempo

```python
n_iteraciones = int(0.15/dt) + 1
t = linspace(0,0.15,n_iteraciones)
```

Se crea el arreglo u que guardara los resultados de la integracion y se inicializa con la condicion inicial.

```python
u = zeros([n_iteraciones,200])
```

```python
u[0,:] = IC
```

Se realiza la integracion de la ecuacion con diferencias centradas

```python
for j in range(n_iteraciones-1):
    u[j+1,0] = 0
    u[j+1,199] = 0
    for i in range(1,199):
        u[j+1,i] = u[j,i] - beta*(u[j,i+1]**2 - u[j,i-1]**2)/2.0; 
```

Se grafica la solucion

```python
figura = plt.figure()
axis = figura.add_subplot(1, 1, 1, projection='3d')
X,Y = meshgrid(t, x)
p = axis.plot_surface(X, Y, transpose(u))
```

Lax-Wendroff

```python
dx = 2.0/199
dt = 0.001
beta = dt/dx
n_iteraciones = int(0.15/dt)
t = linspace(0,0.15,n_iteraciones)
x = linspace(0,2,200)
IC = 3.0*sin(pi*x)
u = zeros([n_iteraciones,200])
u[0,:] = IC
```

El cambio respecto al caso anterior se da en la integracion.

```python
for j in range(n_iteraciones-1):
    u[j,0] = 0
    u[j,199] = 0
    for i in range(1,len(u[j,:]) - 1):
        u[j+1,i] = u[j,i] - beta*(u[j,i+1]**2 - u[j,i-1]**2)/4.0 + ((u[j,i+1] + u[j,i])*(u[j,i+1]**2 - u[j,i]**2))*beta**2/8.0 - (u[j,i] + u[j,i-1])*(u[j,i]**2 - u[j,i-1]**2);
```

```python
figura = plt.figure()
axis = figura.add_subplot(1, 1, 1, projection='3d')
X,Y = meshgrid(t, x)
p = axis.plot_surface(X, Y, transpose(u))
```

Ambos metodos arrojan soluciones bastante distintas al problema, aunque ambos coinciden en una divergencia debido a que la derivada es muy pronunciada en algunas partes de la funcion.
