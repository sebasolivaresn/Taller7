```python
%pylab inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
x = linspace(0,2,200)
IC = 3.0*sin(pi*x)
```

```python
dx = 2.0/199
dt = 0.001
beta = dt/dx
```

```python
n_iteraciones = int(0.15/dt)
print n_iteraciones
t = linspace(0,0.15,n_iteraciones)
```

```python
u = zeros([200,n_iteraciones])
print size(t), size(x), shape(u)
```

```python
u[:,0] = IC
```

```python
for i in range(1,n_iteraciones):
    u[0,i] = 0
    u[199,i] = 0
    for j in range(1, len(u[:,i]) - 1):
        u[j,i] = u[j,i-1] - beta*(u[j+1,i-1]**2 - u[j-1,i-1]**2)/2.0
```

```python
def solucion(x,y):
    return x+y
```

```python
figura = plt.figure()
axis = figura.add_subplot(1, 1, 1, projection='3d')
X,Y = meshgrid(x, t)
Z = solucion(X,Y)
p = axis.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)
```
